```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
df = pd.read_csv(url,sep="\t")
df.head()
```

```python
liste = []
def dolarlastir(df):
    for i in df['item_price']:
        liste.append(float(i[1:-1]))
        return liste

dolarlastir(df)
```

### Step 4. How many products cost more than $10.00?

```python
liste = []
for i in df['item_price']:
    liste.append(float(i[1:-1]))
    
df['item_price'] = liste
```

```python
df.head()
```

```python
#FANCY INDEXING

ondan_buyuk_veriler = df[df['item_price']>10]
ondan_buyuk_veriler[['item_name','item_price']]
```

```python
filtrelenmis_veri = df[df['quantity']==1]
```

```python
son_veri = filtrelenmis_veri[filtrelenmis_veri['item_price']>10]
```

```python
son_veri = son_veri.sort_values(by = 'item_price',ascending=False)
```

```python
last = son_veri[['item_name','choice_description','item_price']]
```
